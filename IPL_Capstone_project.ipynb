```python
ipl['highest_wicket']=ipl['best_bowling_figure'].apply(lambda x:x[0])
ipl
```

```python
top_bowlers=ipl.groupby('best_bowling')['highest_wicket'].sum().sort_values(ascending=False).head(10)
top_bowlers
```

```python
sns.barplot(y=top_bowlers.index,x=top_bowlers.values)
```
